# Bernoulli and Gaussian Naive Bayes Algorithm 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB,GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
df=pd.read_csv("bank.csv")
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df[['age','balance','duration']]=StandardScaler().fit_transform(df[['age','balance','duration']])
```

```python
df.head()
```

```python
x=df.iloc[:, :-1]
y=df.iloc[:, -1]
```

```python
type(y)
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=21,test_size=0.20)
```

```python
def model(M):
    M.fit(x_train,y_train)
    predte=M.predict(x_test)
    print(classification_report(y_test,predte))
    return M
          
```

## Bernoulli Naive Bayes

```python
model(BernoulliNB())
```

## Gaussian Naive Bayes

```python
model(GaussianNB())
```

## Gaussian Model is providing more accurate results for this dataset.
